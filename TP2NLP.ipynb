```python
import os


file_path = os.path.join(os.path.expanduser("~"), "Downloads", "text")
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()
print(text)
```

```python
import string
def clean_text(text):
    # delete the punctiation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # convert to lowercase
    text = text.lower()
    return text
cleaned_text = clean_text(text)
print(cleaned_text)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
tokenized_text = word_tokenize(cleaned_text)
```

```python
print(tokenized_text)
```

```python
from gensim.models import Word2Vec
model = Word2Vec([tokenized_text], vector_size=100, window=5, min_count=1, workers=4)
```

```python
word_representations = {}
for word in model.wv.key_to_index:
     word_representations[word] = model.wv[word]
print(word_representations)
```

```python
similarity_score = model.wv.similarity('marrakech', 'city')
print(similarity_score)
```

```python
similarity_score = model.wv.similarity('marrakech', 'marrakech')
print(similarity_score)
```

```python
similarity_score = model.wv.similarity('marrakech', 'medina')
print(similarity_score)
```
